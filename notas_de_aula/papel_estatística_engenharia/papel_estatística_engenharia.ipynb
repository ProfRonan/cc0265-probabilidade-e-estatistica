### Exemplo (Força de Tração de um Cabo)

Como saber quanto um cabo aguenta tracionar um objeto baseado na sua largura?

Um experimento realizado mostra dois tipos de largura de cabos como pode ser visto na figura.
Qual seria o melhor para garantir uma força de pelo menos 13?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
data_A = {'value': [11.8, 12.5, 12.5, 13.0, 13.1,
                    13.5, 13.7, 13.7, 13.9, 14.2], 'group': ['A'] * 10}
data_B = {'value': [11.9, 12.2, 12.5, 13.1, 13.2,
                    13.7, 13.8, 13.9, 13.9, 14.0], 'group': ['B'] * 10}
df_A = pd.DataFrame(data_A)
df_B = pd.DataFrame(data_B)
df = pd.concat([df_A, df_B]).reset_index(drop=True)
ax = sns.catplot(data=df, kind="swarm", x='value',
                 hue='group', s=100, height=2, aspect=2)
ax.set(title="Valor da tração máxima por grupo", xlabel='valor')
plt.show(ax)
```

## População e Amostra

As vezes é impossível ou impraticável analisar **todos** os elementos de um conjunto de dados que queiramos estudar.

Nesse caso, podemos tentar analisar somente uma parte dos dados (_amostra_) e utilizar as medidas da amostra para tratar de forma geral o todo (_população_).

Esse tipo de técnica é chamada _inferência estatística_.

Obviamente, isso pode gerar erros chamados de _erros amostrais_.
Mas, podemos usar técnicas de amostragem para conseguir uma amostra que represente _"bem"_ a população.

## Coleta de Dados

As vezes, os dados para se trabalhar são a totalidade dos dados.
Nesse caso, temos um _censo_.

Mas na grande maioria das vezes, temos somente uma _amostra_.

- **Estudo Retrospectivo**: Utilizar dados históricos já coletados por outra pessoa/pesquisador.
- **Estudo Observacional**: Observar a população ou fenômeno tentando perturbá-la(o) o mínimo possível.
- **Experimentos**: Reproduzir o fenômeno de forma controlada, podendo alterar parâmetros do experimento de forma sistemática.


## Observação de Dados ao Longo do Tempo

Para dados coletados ao longo do tempo, normalmente, é de grande importância visualizar o seu comportamento por um gráfico dessa _série temporal_.


```python
# Dados sobre temperatura e piratas
data = {'year': [1820, 1860, 1880, 1920, 1940, 1980, 2000],
        'average_temperature': [14.25, 14.35, 14.75, 14.85, 15.25, 15.65, 15.95],
        'num_pirates': [45000, 35000, 20000, 15000, 5000, 400, 17]}
df = pd.DataFrame(data)

# gráfico da temperatura ao longo dos anos
ax = sns.lineplot(data=df, x='year', y='average_temperature')
ax.set(title="Temperatura ao longo dos Anos",
       xlabel='ano', ylabel='temperatura média em °C')
plt.show()

# gráfico dos piratas ao longo dos anos
ax = sns.lineplot(data=df, x='year', y='num_pirates')
ax.set(title="Piratas ao longo dos Anos",
       xlabel='ano', ylabel='quantidade de piratas')
plt.show()
```

## Modelos Mecanicistas e Empíricos

Um _modelo mecanicista_ é construído a partir do nosso conhecimento de mundo.

A _corrente elétrica_ pode ser medida com
$$\operatorname{Corrente} = \operatorname{Voltagem}/\operatorname{Resistencia}$$
ou
$$I = E/R.$$

Contudo ao se medir empiricamente esses valores em um experimento, vamos observar que a relação é quase essa mais ou menos um $\epsilon$.

$$I = E/R + \epsilon$$

Um modelo empírico, na realidade, é observado na forma reversa...
Fazendo-se várias medidas de $I$, $E$ e $R$ e com alguma intuição/criatividade poderíamos encontrar que $I \approx E/R$ (quase).
